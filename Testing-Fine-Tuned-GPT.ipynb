```python
import pandas as pd
import json
from openai import OpenAI
from sklearn.metrics import accuracy_score

client = OpenAI(api_key='sk-proj-*********************************')

# Load dataset
file_path = '/content/preprocessed_combined_file (1).xlsx'
df = pd.read_excel(file_path)

# Sample 50 examples from the dataset
df_sampled = df.sample(n=50, random_state=42)

# Function to get predictions from the fine-tuned model
def get_prediction(text, model_id):
    response = client.chat.completions.create(
        model=model_id,
        messages=[
            {"role": "system", "content": "Classify the sentiment of this Text."},
            {"role": "user", "content": text}
        ]
    )
    # Access the content from the ChatCompletionMessage object
    return response.choices[0].message.content

# Generate predictions for the sampled examples
model_id = 'ft:gpt-3.5-turbo-0125:comsats:hate-speech:9njGZpq7'
predictions = []
for Text in df_sampled['Text']:
    prediction = get_prediction(Text, model_id)
    predictions.append(int(prediction))  # Ensure the prediction is in the correct format (integer)

# Compute accuracy
accuracy = accuracy_score(df_sampled['Tag'], predictions)
print(f"Model Accuracy on 50 examples: {accuracy * 100:.2f}%")

# Display predictions and actual labels
df_sampled['Predicted Tag'] = predictions
print(df_sampled[['Text', 'Tag', 'Predicted Tag']])
```
